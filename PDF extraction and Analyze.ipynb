```python
import camelot                           #import camelot library to use 'read_pdf' module
import pandas as pd
```

```python
table = camelot.read_pdf('Marks.pdf')     #get the 'Marks.pdf' file and assign it to a variable
table = table[0].df                       #get the data frame of the pdf file
table
```

```python
table = table.drop([table.index[0],table.index[12]])   #remove unwanted rows
table.replace(True)
table
```

```python
table.drop([0,5],inplace=True,axis=1)  # remove unwanted columns
table
```

```python
new_col_names = ['Name','Mathematics','Science','English']   # Add a header names to the columns 
table.columns = new_col_names                                # Assign the new columns name to the data frame
table
```

```python
table = table.drop([table.index[0]]) # remove the unwanted row  
table.inplace=True     # dataframe will return nothing and the dataframe is now updated
table
```

```python
table.dtypes    # get the type of columns.All the columns are in 'object' type.Therefore unable to do calculations
```

```python
table['Mathematics'] = table['Mathematics'].astype(int)      #convert the  data types in object to integers 
table['Science'] = table['Science'].astype(int)
table['English'] = table['English'].astype(int)
```

```python
table.dtypes   # Again check the data types of columns "Mathematics", "Science" , "English" columns data types have been updated 
```

```python
columns_list = ['Mathematics', 'Science','English']             # make a list of columns which we want to get sum
table['Total'] = table[columns_list].sum(axis=1)                # get the total marks of each students 
table
```

```python
table.reset_index(drop=True, inplace=True)  # reset the index values
table
```

```python
table.to_csv("Rangana.csv")  #Export the CSV file to local computer 
table
```
